# Análise e limpeza de dados
## Introdução
Nesse projeto vamos analisar o dataset de medalhistas olímpicos, vamos começar importando as bibliotecas que serão utilizadas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Agora vamos importar nosso dataset

```python
atletas = pd.read_csv("df_athlete_events.csv")
atletas.head()
```

Para entender melhor nosso dataset, devemos extrair algumas informações sobre ele

```python
atletas.info()
```

Podemos notar uma inconsistência, que se encontra na coluna "Weight" que apresenta o dtype errado, como os valores dessa coluna representam o peso de cada atleta devemos converter ela para "float64".

```python
# antes de converter o tipo da coluna, vamos analisar quais são os dados presentes nas colunas
atletas["Weight"].unique()
```

```python
# vamos substituir alguns valores que não são coerentes com a coluna
atletas["Weight"].replace("hsuahusahu", np.nan, inplace = True)
```

```python
# agora que todos os valores podem ser interpretados como numéricos, vamos converter o tipo da coluna
atletas["Weight"] =atletas["Weight"].astype("float64") 
```

```python
atletas.info()
```

## Dados categóricos
Vamos analisar agora os dados categóricos, garantindo que valores iguais apresentem o mesmo formato. Para as colunas "Name", "Team" e "Event", não será realizada a limpeza, pois nelas há diversos valores possíveis, o que dificulta a análise se os dados são consistentes ou não. Além disso, essas colunas não apresentam grande potencial para a análise dos dados

### Sex

```python
atletas["Sex"].unique()
```

```python
#Vamos substituir dados coerentes por sua representação unificada, e dados incoerentes por NaN
atletas["Sex"].replace("MALE", "M", inplace = True)
atletas["Sex"].replace("FEMALE", "F", inplace = True)
atletas["Sex"].replace("Vic vai te dar um socao", np.nan, inplace = True)
atletas["Sex"].replace("P&D >> Comunidade", np.nan, inplace = True)
```

```python
atletas["Sex"].unique()
```

### NOC

```python
atletas["NOC"].unique()
```

Podemos notar que todas as siglas seguem um padrão bem definido.

### Games

```python
atletas["Games"].unique()
```

### Season

```python
atletas["Season"].unique()
```

```python
#Algumas colunas apresentam informações relevantes, mas devemos substituí-las para uma melhor análise dos dados
atletas["Season"].replace("#ForaBolsonaro", np.nan, inplace = True)
atletas["Season"].replace("Ele nunca", np.nan, inplace = True)
atletas["Season"].replace("Ele não", np.nan, inplace = True)
```

```python
atletas["Season"].unique()
```

### City

```python
atletas["City"].unique()
```

```python
#Devemos substituir os dados que não correspondem a uma cidade
atletas["City"].replace("######", np.nan, inplace = True)
```

```python
atletas["City"].unique()
```

### Sport

```python
atletas["Sport"].unique()
```

### Medal

```python
atletas["Medal"].unique()
```

Antes de substituir os valores incoerentes é importante substituir os NaN para outro valor, pois eles representam atletas que não ganharam medalhas 

```python
atletas["Medal"].replace(np.nan, "Sem medalha", inplace = True)
```

```python
atletas["Medal"].replace("Nelson", np.nan, inplace = True)
```

```python
atletas["Medal"].unique()
```

## Dados numéricos
Agora vamos analisar os dados numéricos, conferindo se seus valores são consistentes.

### Age

```python
atletas["Age"].describe()
```

Tudo correto para essa coluna

### Height

```python
atletas["Height"].describe()
```

A altura máxima está bem elevada, vamos conferir esse valor

```python
#Vamos filtrar por atletas que tenham altura maior do que 5 metros
atletas[atletas["Height"]>500]
```

Vamos substituir esses valores irreais por NaN's

```python
atletas["Height"].describe()
```

Muito melhor

### Weight 

```python
atletas["Weight"].describe()
```

Tudo certo nessa coluna

### Year

```python
atletas["Year"].describe()
```

Podemos notar que há um valor irreal no mínimo, vamos substituí-lo

```python
atletas["Year"] = atletas["Year"].mask(atletas["Year"]<=1800, np.nan)
```

```python
atletas["Year"].describe()
```

Agora sim, bem melhor!!

## Análise dos dados

Primeiro será analisado o número de participantes divididos pelo sexo

```python
fig, ax = plt.subplots(figsize = (11, 6))
sns.countplot(x = atletas["Sex"].dropna(), ax = ax) 
plt.xlabel("Sexo")
plt.ylabel("Número de total de participantes")
plt.title("Comparação do número de participantes divididos pelo sexo")
plt.show()
```

Nota-se que há um número muito maior de participantes masculinos em comparação às participantes femininas.
Será analisado então, como essa diferença se deu com o tempo

```python
fig, ax = plt.subplots(figsize = (11, 6))
sns.countplot(hue = atletas["Sex"], x = atletas["Year"].dropna().astype("int64"), ax = ax,  palette = "Greys") 
plt.xlabel("Ano")
plt.ylabel("Número de participantes")
plt.title("Comparação do número de homens e mulheres nas olimpíadas conforme o tempo")
plt.legend(loc = "upper left")
plt.xticks(rotation = 90)
plt.show()
```

Para melhorar nossa análise, vamos separar os gráficos entre as olimpíadas de inverno e de verão, que apresentam números bem distintos de participantes

```python
fig, ax = plt.subplots(figsize = (11, 6))
winter = atletas[atletas["Season"] == "Winter"]
sns.countplot(hue = winter["Sex"], x = winter["Year"].dropna().astype("int64"), ax = ax, palette = "Blues") 
plt.xlabel("Ano")
plt.ylabel("Número de participantes")
plt.title("Comparação do número de homens e mulheres nas olimpíadas de inverno conforme o tempo")
plt.legend(loc = "upper left")
plt.xticks(rotation = 90)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (11, 6))
summer = atletas[atletas["Season"] == "Summer"]
sns.countplot(hue = summer["Sex"], x = summer["Year"].dropna().astype("int64"), ax = ax, palette = "Oranges") 
plt.xlabel("Ano")
plt.ylabel("Número de participantes")
plt.title("Comparação do número de homens e mulheres nas olimpíadas de verão conforme o tempo")
plt.legend(loc = "upper left")
plt.xticks(rotation = 90)
plt.show()
```

De maneira geral, podemos notar a diferença absurda entre o número de homens e mulheres nas edições iniciais. Essa diferença diminuiu ano a ano, porém, até os dias atuais ainda há um maior número de participantes masculinos, tanto nas olimpíadas de inverno quanto nas de verão. 

Além disso, é notável a diferença no número de participantes das olimpiadas de inverno em relação com as olimpíadas de verão. Vamos nos aprofundar mais nas diferenças entre as duas.

```python
fig, ax = plt.subplots(figsize=(11, 6))
winter_noc = winter.groupby("NOC")["Athlete_ID"].count().sort_values()[::-1][:20]

sns.barplot(x = winter_noc.index, y = winter_noc.values, ax = ax, palette = "Blues_r")
plt.title("Comitês com os maiores números de atletas nas olimpíadas de inverno")
plt.xlabel("Comitê olimpíco")
plt.ylabel("Número de atletas")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(11, 6))
summer_noc = summer.groupby("NOC")["Athlete_ID"].count().sort_values()[::-1][:20]

sns.barplot(x = summer_noc.index, y = summer_noc.values, ax = ax, palette = "Oranges_r")
plt.title("Comitês com os maiores números de atletas nas olimpíadas")
plt.xlabel("Comitê olimpíco")
plt.ylabel("Número de atletas")
plt.show()
```

Os EUA aparecem como o país que mais enviou atletas para as olímpiadas de inverno e verão. Mas é interessante notar países como Canadá, Noruega, Finlândia e Suíça aparecem muito mais nas olimpíadas de inverno em comparação com as de verão, mostrando que os participantes de cada uma

Vamos nos aprofundar agora na análise dos medalhistas.

```python
fig, ax = plt.subplots(figsize = (11, 6))
sns.countplot(x = "Medal", data = atletas, ax = ax, order = ["Gold", "Silver", "Bronze", "Sem medalha"])
ax.set_title("Comparação do número de atletas que ganharam medalha")
```

Como esperado, o número de ouros, pratas e bronzes é praticamente igual. Mas é interessante notar o baixo número de atletas que não ganharam medalhas e estão presentes no nosso dataset.

```python
import warnings

medalhistas = atletas[atletas["Medal"].isin(["Gold", "Silver", "Bronze"])]
with warnings.catch_warnings():
    warnings.simplefilter('ignore')
    medalhistas["Ouro"] = medalhistas["Medal"] == "Gold"
    medalhistas["Prata"] = medalhistas["Medal"] == "Silver"
    medalhistas["Bronze"] = medalhistas["Medal"] == "Bronze"
medalhas_por_pais = medalhistas.groupby("NOC")[["Ouro", "Prata", "Bronze"]].sum()
medalhas_ordenadas = medalhas_por_pais.sort_values(["Ouro", "Prata", "Bronze"], ascending = False)
top_10 = medalhas_ordenadas[:10]
```

```python
fig, ax = plt.subplots(figsize = (11, 6))
ax.bar(x = top_10.index, height = top_10["Ouro"], color = "#FFD700", label = "Ouro")
ax.bar(x = top_10.index, height = top_10["Prata"], color = "#C0C0C0", label = "Prata", bottom = top_10["Ouro"])
ax.bar(x = top_10.index, height = top_10["Bronze"], color = "#CD7F32", label = "Bronze", bottom = top_10["Ouro"] + top_10["Prata"])
plt.legend()
ax.set_title("Comitês olimpícos que mais ganharam medalhas")
ax.set_xlabel("Comitê olimpíco")
ax.set_ylabel("Número de medalhas")
plt.show()
```
